# Overview

The purpose of the Spectral Parameters Toolkit is to investigate the ability of a given instrument to discriminate between defined groups of materials, by sampling of the material reflectance spectra at the instrument resolution.

Here we demonstrate the steps of defining a set of filter centre-wavelengths and bandwidths, and use these to cosntruct the high-resolution spectral transmission profiles of the channels, used to compute the expected sampled reflectance of the given MaterialCollection.

The steps are:
1. Define the instrument channel centre-wavelengths and bandwidths, and write these to file.
2. Build the instrument from the channel definition file.
3. Produce a visualisation of the instrument channel transmission profiles.

# Defining the Instrument Channels

Here we will use the ```InstrumentBuilder``` function to build an instrument with 8 channels, distributed from 400 nm to 1000 nm, with a resolution of $\lambda / \Delta\lambda = 5$.

````python
from sptk.instrument import InstrumentBuilder
````

````python
cwls = [440.0, 500.0, 580.0, 650.0, 780.0, 850.0, 950.0, 1000.0]
inst = InstrumentBuilder('example_instrument', 'filter-wheel', cwls, 10.0)
inst.build_instrument()
````

Now that we have written the table of channel properties, we can read this with the ```Instrument``` constructor function. It is also possible to build a simple csv file of the required information by hand, for example, if bespoke bandwidths need be defined, that cannot simply be summarised by the $\lambda / \Delta\lambda$ parameter.

# Building the Instrument

We now import the ```Instrument``` class and build the instrument transmission profiles at high spectral resolution by calling the class constructor.

````python
from sptk.instrument import Instrument
````

We'll use the same ```'sptk_example``` project name as previously.

````python
inst = Instrument(
            name = 'example_instrument',
            project_name = 'sptk_example',
            load_existing=False)
````

We can now see the transmission profiles produced. The transmission data has been exported to the ```example_spectral_library``` directory, where it can be accessed independently to this notebook.
